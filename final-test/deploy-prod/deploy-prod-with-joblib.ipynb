# Deploy in Prod

The trained pipeline was serialized using joblib and deployed as a single artifact, enabling consistent preprocessing and inference in production.

```python
import joblib
import pandas as pd

model = joblib.load("heart_disease_pipeline.joblib")

new_patient = pd.DataFrame([{
    'age': 55, 'trestbps': 130, 'chol': 245, 'thalach': 150,
    'oldpeak': 1.0, 'ca': 0, 'sex': 1, 'cp': 1, 'fbs': 0,
    'restecg': 1, 'exang': 0, 'slope': 2, 'thal': 2
}])

pred = model.predict(new_patient)

print("Patient Sick" if pred[0] == 1 else "Healthy")
```
